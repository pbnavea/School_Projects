```python
import itertools
import inspect

def equivalence(a,b):
    return a == b

def truthTable(f):
    #lenght of arguments available(?)
    n = len(inspect.signature(f).parameters) 
    x = list(itertools.product([True,False],repeat =n)) 
    
#saves to the "x" variable the list of possible True/False compinations
#"n" gives the number 2^n
    return x
```
